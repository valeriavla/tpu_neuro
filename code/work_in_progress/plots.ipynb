```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the datasets and their labels
datasets = [nlp_random_10, nlp_random_50, nlp_random_100, nlp_default_10, nlp_default_50, nlp_default_100, xla_random_10, xla_random_50, xla_random_100, xla_default_10, xla_default_50, xla_default_100]
labels = ['NLP_Random_10', 'NLP_Random_50','NLP_Random_100','NLP_Default_10','NLP_Default_50','NLP_Default_100','XLA_Random_10','XLA_Random_50','XLA_Random_100','XLA_Default_10','XLA_Default_50','XLA_Default_100']

# Creating combined plots
plt.figure(figsize=(12, 32))

for i, dataset in enumerate(datasets):
    label = labels[i]
    
    # Plot for Losses (Training and Validation)
    plt.subplot(12, 2, i * 2 + 1)
    plt.plot(dataset['epoch'], dataset['training_loss'], label='Training Loss', marker='o', color='purple')
    plt.plot(dataset['epoch'], dataset['validation_loss'], label='Validation Loss', marker='o', color='orange')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(f'{label} Training vs Validation Loss')
    plt.legend()
    plt.xticks(dataset['epoch'].astype(int)) 
    
    # Plot for OPAs (Training and Validation)
    plt.subplot(12, 2, i * 2 + 2)
    plt.plot(dataset['epoch'], dataset['training_opa'], label='Training OPA', marker='o', color='purple')
    plt.plot(dataset['epoch'], dataset['validation_opa'], label='Validation OPA', marker='o', color='orange')
    plt.xlabel('Epoch')
    plt.ylabel('OPA')
    plt.title(f'{label} Training vs Validation OPA')
    plt.legend()
    plt.xticks(dataset['epoch'].astype(int)) 


plt.tight_layout()
plt.savefig('../plots/final_metric_plots.png')
plt.show()
```
